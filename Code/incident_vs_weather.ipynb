```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Store filepath in a variable
weather_file= "Data/CentralPark15-17.csv"
flu_incident_file= "Data/ILINet_2015_2017.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
weather_file_df = pd.read_csv(weather_file, encoding = "ISO-8859-1")
```

```python
weather_file_df
```

```python
# organized_df = training_data[["Name","Trainer","Weight","Membership(Days)"]]
# organized_df.head()
weath_summ_df = weather_file_df[["PRCP", "SNOW", "TAVG", "WkStart"]]
weath_summ_df.head()
```

```python
incident_file_df = pd.read_csv(flu_incident_file, encoding = "ISO-8859-1")
```

```python
incident_file_df
```

```python
inc_summ_df = incident_file_df[["WkStart","ILITOTAL" ]]
inc_summ_df.head()
```

```python
# merge_table = pd.merge(info_pd, items_pd, on="customer_id")
inc_weath_df = pd.merge(inc_summ_df, weath_summ_df, on ="WkStart")
inc_weath_df.head()
```

```python
x_axis =  np.arange(len(inc_weath_df))
x_axis
```

```python
# xticks(arange(len(inc_weath_df)),inc_weath_df["WkStart"], rotation=17)
# Avearge weather temp
points_F = inc_weath_df["TAVG"]
```

```python
inc_vol = inc_weath_df["ILITOTAL"]
```

```python
# Create a handle for each plot
temperature, = plt.plot(x_axis, points_F, marker="+",color="blue", linewidth=1, label="Temperature")
incidents, = plt.plot(x_axis, inc_vol, marker="s", color="Red", linewidth=1, label="Incidents")
```

```python
# Set our legend to where the chart thinks is best
plt.legend(handles=[temperature, incidents], loc="best")
```

```python
# Create labels for the X and Y axis
plt.xlabel("Weeks")
plt.ylabel("Degrees/Incidents")
```

```python
# Save and display the chart
plt.savefig("temp_vs_incidents.png")
plt.show()
```

```python
# decent start.  Let's build another one, this time with separate y axes.  

fig, ax1 = plt.subplots()               # fig, ax1 = plt.subplots()
xshared = np.arange(len(inc_weath_df))  # t = np.arange(0.01, 10.0, 0.01)
y1_points_F = inc_weath_df["TAVG"]      # s1 = np.exp(t)
ax1.plot(xshared, y1_points_F, 'b-')    #  ax1.plot(t, s1, 'b-')
ax1.set_xlabel('Weeks')                 # ax1.set_xlabel('time (s)')

# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('Temperature (F)', color = 'b')  # ax1.set_ylabel('exp', color='b')

ax1.tick_params('y', colors = 'b')      # ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()                       # ax2 = ax1.twinx()
y2_inc_vol = inc_weath_df["ILITOTAL"]   # s2 = np.sin(2 * np.pi * t)
ax2.plot(xshared, y2_inc_vol, 'r.')     #  ax2.plot(t, s2, 'r.')
ax2.set_ylabel ('Flu Incidents', color = 'r')  #  ax2.set_ylabel('sin', color='r')
ax2.tick_params ( 'y', colors = 'r')      #  ax2.tick_params('y', colors='r')

fig.tight_layout()
plt.show()
```
